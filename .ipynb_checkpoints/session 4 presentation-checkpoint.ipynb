# Virtual enviroment (venv)
  is a tool to create isolated Python environments.Normally pip and venv arre included in python 

## Installation
  ### via pip
    
    pip is a de facto standard package-management system used to install and 
    manage software packages written in Python
  
  #### Install Pip(windows)
    -via Cmd check first if pip is installed or not 
      python -m pip --version
    -To install pip, securely  download get-pip.py by following this link: 
      curl https://bootstrap.pypa.io/get-pip.py -o get-pip.py
      python get-pip.py 
    -check pip version
      python -m pip --version
  #### install Venv

      python -m pip install  virtualenv
      python -m virtualenv --help


## Create Virtual env 
    -via CMD, VS code, Pycharm, etc 
      cd project_folder
      python -m venv ./venv (will be created in the same folder 
    -or 
      virtualenv venv
## Activate Virtual env 
    -via CMD, VS code, Pycharm, etc 
       ./venv/Scripts/activate.bat
    -or 
        python path/to/project/file/venv/Scripts/activate.bat

## Deactivate Virtual env 
    -via CMD, VS code, Pycharm, etc 
      deactivate


# Django
    Django is a Python-based free and open-source web framework that follows the model-template-views architectural pattern.

## Insatll (prefer Virtual env)
    -via CMD, VS code, Pycharm, etc 
      pip install Django

## Create Django Project
    -via CMD, VS code, Pycharm, etc 
      django-admin.py startproject PROJECT(name of project)

## Start Django
    -via CMD, VS code, Pycharm, etc 
      cd PROJECT
      python3 manage.py startapp APP

# Common Django Files 

## Setting.py
    include most application settings 
    
##### SECRET_KEY 
      -There are many things in a Django app which require a cryptographic signature that what the key is used 
      for 
        SECRET_KEY = 'k+f+#0+j(d$+-x%oj#x_$3(s$!54kbnc$iztk)i1=w_y-jc28k'

##### DEBUG 
      -it display code warnings and errors in the development phase  
           DEBUG = True


#### ALLOWED_HOSTS
      -it is used in deployment phase to allow access to users
        ALLOWED_HOSTS = []
  

##### INSTALLED_APPS
      -contains a registry of installed applications that stores configuration and provides introspection. It also maintains a list of available models.
    INSTALLED_APPS = [
    'crispy_forms',
    'pages.apps.PagesConfig',
    'bookupload.apps.BookuploadConfig',
    'authen.apps.AuthenConfig',
    'django.contrib.admin',
    'django.contrib.auth',
    'django.contrib.contenttypes',
    'django.contrib.sessions',
    'django.contrib.messages',
    'django.contrib.staticfiles',
    ]


##### TEMPLATES
  contaions the path of html files dirctory
    TEMPLATES = [
    {
        'BACKEND': 'django.template.backends.django.DjangoTemplates',
        'DIRS': [os.path.join(BASE_DIR , 'templates')],
        'APP_DIRS': True,
        'OPTIONS': {
            'context_processors': [
                'django.template.context_processors.debug',
                'django.template.context_processors.request',
                'django.contrib.auth.context_processors.auth',
                'django.contrib.messages.context_processors.messages',
            ],
        },
    },
    ]

##### DATABASES
      contains data base setting like username, password, database name.
            DATABASES = {
                'default': {
                    'ENGINE': 'django.db.backends.sqlite3',
                    'NAME': os.path.join(BASE_DIR, 'db.sqlite3'),
                }
            }

##### Static
      conating path to static files (images, videos, files) which is used in the front-end
        
        STATIC_ROOT = os.path.join(BASE_DIR, 'static')
        STATIC_URL = '/static/'
##### media
    contain the path to files saved in server
    
        MEDIA_ROOT = os.path.join(BASE_DIR, 'media')
        MEDIA_URL = '/media/'

# POST vs GET request 

        GET is used to request data from a specified resource.
        
        
        POST is used to send data to a server to create/update a resource.
